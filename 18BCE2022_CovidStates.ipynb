```python
import pandas as pd
import seaborn as sns
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
fp = r'indian-states.shp'
map_df = gpd.read_file(fp)
map_df.head()
```

```python
data_df = pd.read_excel(r'statewise-dataset.xlsx')
data_df.head()
```

```python
merged = map_df.set_index('st_nm').join(data_df.set_index('Name of State / UT'))
merged.head()
```

```python
fig, ax = plt.subplots(1, figsize=(10, 10))
ax.axis('off')
ax.set_title('Total Confirmed cases in 2020', fontdict={'fontsize': '15'})

# plot the figure
merged.plot(column='Total Confirmed cases',cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0', legend=True, markersize=[39.739192, -104.990337])
```
