```python
import pandas as pd
import seaborn as sns
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
file_path = r'Dist.shp'
df_map = gpd.read_file(file_path)
df_map.head()
```

```python
df_map.plot()
```

```python
df_data = pd.read_csv(r'district.csv')
df_data.head()
```

```python
merged = df_map.set_index('DISTRICT').join(df_data.set_index('District'))
merged.head()
```

```python
fig, ax = plt.subplots(1, figsize=(10, 10))
ax.axis('off')
ax.set_title('Total Confirmed cases in 2020', fontdict={'fontsize': '15'})

merged.plot(column='Confirmed',cmap='RdYlGn_r', linewidth=0.8, ax=ax, edgecolor='0', legend=True, markersize=[39.739192, -104.990337])
```
